```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']

df = pd.read_excel('耦合协调度.xlsx')

for i in range(len(df["地区"])):
    plt.bar(df["地区"][i],df["环境治理效率均值"][i])
# plt.xticks(df['地区'])
plt.xlabel("地区")
plt.ylabel("环境治理效率均值")
```
